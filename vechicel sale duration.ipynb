```python
import matplotlib.pyplot as plt

# Define the years and sales data
years = list(range(2010, 2025))
sales = [
    50000,  # 2010
    75000,  # 2011
    100000, # 2012
    150000, # 2013
    200000, # 2014
    500000, # 2015
    700000, # 2016
    800000, # 2017
    1000000, # 2018
    2200000, # 2019
    3100000, # 2020
    6500000, # 2021
    10000000, # 2022
    12500000, # 2023
    15000000  # 2024
]

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(years, sales, marker='o', linestyle='-', color='b')

# Add titles and labels
plt.title('Global Electric Vehicle Sales (2010-2024)')
plt.xlabel('Year')
plt.ylabel('Sales (Units)')
plt.grid(True)

# Show the plot
plt.xticks(years) # Ensure all years are marked on x-axis
plt.show()
```
